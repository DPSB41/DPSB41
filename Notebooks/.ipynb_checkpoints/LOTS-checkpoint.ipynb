```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
pd.set_option('display.float_format', lambda x: '%.0f' % x)
```

```python
try:
    engine = create_engine("mysql+mysqlconnector://root:a21305556699@localhost:3306/DPSB", echo=False)
    query = """
    SELECT Localisation, MO, COUNT(*) AS "COUNT",SUM(APA) AS "SUM"
    FROM OPERATIONS
    WHERE Localisation IS NOT NULL
    GROUP BY Localisation, MO
    """
    df = pd.read_sql(query, engine)
except Exception as e:
    engine.close()
    print(str(e))
df
```

```python
df["LOT"] = df["SUM"] / (df["Localisation"].str.count(" ") + 1)
```

```python
locals = set()
for local in df["Localisation"]:
    for loc in local.split():
        locals.add(loc)
```

```python
res = pd.DataFrame()
lots = list()
for loc in sorted(locals):
    res = df[df["Localisation"].str.contains(loc) == True]
    res =res.pivot(index=None, columns="MO", values=["COUNT", "LOT"])
    res.replace(np.nan, 0, inplace=True)
    #res.drop(["Localisation"], axis=1)
    lots.append(res)
```

```python
lots[1].index
```

```python
sum_lot = list()
for lot, loc in zip(lots, sorted(locals)):
    lot_pd = pd.DataFrame(lot)
    sum_lot.append(lot_pd)
    
```

```python
sum_lot[1]["COUNT"]
```

```python
resumes = {"LOC": sorted(locals), "SUM": sum_lot}
```
