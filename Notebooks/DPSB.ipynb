```python
import mysql.connector as connection
import pandas as pd
import numpy as np
```

```python
try:
    db = connection.connect(host="localhost", database="DPSB", user="root", password="a21305556699", use_pure=True)
    query = "SELECT Mo, MICL, COUNT(Mo) AS Number, SUM(APA) AS SUM FROM OPERATIONS GROUP BY MO, MICL;"
    df = pd.read_sql(query, db)
    db.close()
except Exception as e:
    db.close()
    print(str(e))
```

```python
df
```

```python
res = df.pivot(index="Mo", columns="MICL", values=["Number", "SUM"])
```

```python
pd.set_option('display.float_format', lambda x: '%.0f' % x)
res = res.replace(np.nan, 0)
```

```python
res
```

```python
TOTAL_OP = res["Number"].sum(axis=1)
TOTAL_OP.name = "TOTAL OP"
Number = [res["Number"], TOTAL_OP]
Number = pd.concat(Number, axis=1)
Number
```

```python
TOTAL_SUM = res["SUM"].sum(axis=1)
TOTAL_SUM.name = "SUM"
SUM = [res["SUM"], TOTAL_SUM]
SUM = pd.concat(SUM, axis=1)
SUM
```

```python
res = [Number, SUM]
res = pd.concat(res, axis=1)
res
```

```python
TOTAL = res.sum(axis=0)
TOTAL
```

```python
TOTAL.name = "TOTAL"
res.append(TOTAL.T)
```

```python
res.set_axis(["NOP A l'Arret", "NOP Achevée", "NOP En Course", "NOP Gelée", "NOP Non Lancée", "TOTAL OP", "TOTAL A l'Arret", "TOTAL Achevée", "TOTAL En Cours","TOTAL Gelée", "TOTAL Non Lancée", "SUM"], axis=1, inplace=np.True_)
res
```

```python
DC = res["TOTAL Gelée"] + res["TOTAL Non Lancée"]
DC.name = "Dépenses Cumulées"
DC
```

```python
res = [res, DC]
res = pd.concat(res, axis=1)
res
```
